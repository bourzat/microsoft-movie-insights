#### Preparing `box_office_movie_gross`

We will start with the `box_office_movie_gross` DataFrame where the `foreign_gross` column is of type `object` and convert it to `float`.

```python
# Removing ','
box_office_movie_gross["foreign_gross"] = box_office_movie_gross["foreign_gross"].str.replace(',', '')
# Removing '.'
box_office_movie_gross["foreign_gross"] = box_office_movie_gross["foreign_gross"].str.replace('.', '')
```

```python
# convert the column to a float
box_office_movie_gross["foreign_gross"] = box_office_movie_gross["foreign_gross"].astype(float)

# check if column has been converted to a float
box_office_movie_gross.info()
```

We have now succesfuly converted out `foreign_gross` column to a `float`.

We have now converted the column to its correct data type which will allow us to perform numerical operations on it. 

Next, we will drop all rows that don't have values in the `foreign_gross column` and all other rows that don't contain values. There are too many values to replace with the mean which will affect our data. 

```python
# Dropping missing values in the df
box_office_movie_gross.dropna(inplace=True)

# Checking for missing values
box_office_movie_gross.isna().sum()
```

```python
# Checking number of records
len(box_office_movie_gross)
```

To determine how much revenue the movies generated, we will create a new column called `total_gross` that will be the sum of both `foreign_gross` and `domestic_gross`.

To make our analysis easier, we should remove the 0's in both the `domestic_gross` and `foreign_gross` and put it in terms of millions. 

```python
# Creating a new column called `domestic_gross (In millions)`
box_office_movie_gross["domestic_gross (millions)"] = box_office_movie_gross["domestic_gross"] / 1000000

# Dropping the old domestic_gross DataFrame
box_office_movie_gross.drop("domestic_gross", axis=1,inplace=True)
```

```python
# Creating a new column called `foreign_gross (In millions)`
box_office_movie_gross["foreign_gross (millions)"] = box_office_movie_gross["foreign_gross"] / 1000000

# Dropping the old foreign_gross DataFrame
box_office_movie_gross.drop("foreign_gross", axis=1,inplace=True)
```

```python
# Creating a new column total_gross
box_office_movie_gross["total_gross (millions)"] = box_office_movie_gross["domestic_gross (millions)"] + box_office_movie_gross["foreign_gross (millions)"]

box_office_movie_gross.head()
```

Next we will drop the `studio` column since it play no part in the analysis.

```python
# Dropping the studio column 
box_office_movie_gross.drop("studio", axis=1, inplace=True)

box_office_movie_gross.head()
```

We can now investigate the `popular genres`. 

Here we have an Entity Relationship Diagram that we can use to write and retrieve necessary queries from the database.
![ERD](Images/ERD.png)

To be able to determine the most popular genres, we must first define what makes a genre popular. In this case, we would need to find the average of both the numvotes and the averagerating.

```python
# Writing a query to retrieve info from the database
query = """SELECT mb.genres, 
                  mr.averagerating,
                  mr.numvotes
           FROM
                  movie_basics mb
           JOIN
                  movie_ratings mr ON mb.movie_id = mr.movie_id;"""

# Querying db and creating a new df
genres_df = pd.read_sql(query, conn)
```

```python
genres_df.head()
```

```python
# Gaining some info on the columns data types
genres_df.info()
```

Now let's find the mean of the average rating and the mean number of votes.

```python
# Finding mean of the averagerating
genres_df['averagerating'].mean()
```

```python
# Median number of votes
genres_df['numvotes'].mean()
```

Assuming a successful movie gets an average rating of `7.5` and more than `1000` votes, we can filter the DataFrame and find the successful genres.

```python
# Creating a new DF with only successful genres
successful_genres = genres_df.loc[(genres_df['averagerating'] > 7.5) & (genres_df['numvotes'] > 1000)]
```

```python
successful_genres.head()
```

```python
# Finding the unique genres
unique_genres = successful_genres['genres'].value_counts()
unique_genres.head()
```

Let's find the `top 5 genres` in the DataFrame.

```python
# Calculate the average rating for each unique genre combination
average_ratings = successful_genres.groupby('genres')['averagerating'].mean().sort_values(ascending=False)
```

```python
#Top 5 genre combinations by average rating
top_5_genres = average_ratings.head(5)
```

```python
top_5_genres = pd.DataFrame(top_5_genres)
```

```python
# Visualize the top 5 genre combinations
plt.figure(figsize=(10, 6))
top_5_genres.plot(kind='bar')
plt.title('Top 5 Genre Combinations by Average Rating')
plt.xlabel('Genre Combinations')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

The following genre combinations `Comdey, Documentary, History`, `Adventure, Drama, Romance`, `Comedy, Documentary`, `Action, Crime` and `Musical` are the most successful genres and the production team should focus on creating movies around these.
